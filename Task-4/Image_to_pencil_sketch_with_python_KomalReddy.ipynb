# **Task-4 : Image to Pencil Sketch with Python(Beginner Level Task)**
We need to read the image in RBG format and then convert it to a grayscale image. This will turn an image into a classic black and white photo. Then the next thing to do is invert the grayscale image also called negative image, this will be our inverted grayscale image. Inversion can be used to enhance details. Then we can finally create the pencil sketch by mixing the grayscale image with the inverted blurry image. This can be done by dividing the grayscale image by the inverted blurry image. Since images are just arrays, we can easily do this programmatically using the divide function from the cv2 library in Python.

***Images used*** : a car and a butterfly (we will work with both)

## A look into the images

```python
import cv2 #importing library for reading images (OpenCV (Open Source Computer Vision Library) is an open source computer vision and machine learning software library.)
from google.colab.patches import cv2_imshow # cv2.imshow() is disabled in Colab, because it causes Jupyter sessions to crash
image1 = cv2.imread("/content/car im 1.jpg")
cv2_imshow(image1)
```

```python
image2 = cv2.imread("/content/butterfly im 2.jpg")
cv2_imshow(image2)
```

## Converting into a Gray scale image

```python
gray_image1 = cv2.cvtColor(image1, cv2.COLOR_BGR2GRAY) #cv2.cvtColor() method is used to convert an image from one color space to another
cv2_imshow(gray_image1)
```

```python
gray_image2 = cv2.cvtColor(image2, cv2.COLOR_BGR2GRAY) #cv2.cvtColor() method is used to convert an image from one color space to another
cv2_imshow(gray_image2)
```

## Invert the Gray scale called as negative scale

```python
inverted_image1 = 255 - gray_image1
cv2_imshow(inverted_image1)
```

```python
inverted_image2 = 255 - gray_image2
cv2_imshow(inverted_image2)
```

## Blurring the image

```python
blurred1 = cv2.GaussianBlur(inverted_image1, (21, 21), 0) #(21,21)-> size of kernel
cv2_imshow(blurred1)
```

```python
blurred2 = cv2.GaussianBlur(inverted_image2, (21, 21), 0) #(21,21)-> size of kernel
cv2_imshow(blurred2)
```

## Inverting the blurred image and dividing the grayscale image

```python
inverted_blurred1 = 255 - blurred1
cv2_imshow(inverted_blurred1)
```

```python
pencil_sketch1 = cv2.divide(gray_image1, inverted_blurred1, scale=256.0)
cv2_imshow(pencil_sketch1)
```

```python
inverted_blurred2 = 255 - blurred2
cv2_imshow(inverted_blurred2)
```

```python
pencil_sketch2 = cv2.divide(gray_image2, inverted_blurred2, scale=256.0)
cv2_imshow(pencil_sketch2)
```
